```python

def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n - 1)

number = int(input("Enter the number: "))
factorial = fact(number)
print(factorial)



    
```

```python
number=int(input("Enter the number: "))
factorial=1
while number>0:
    factorial=factorial*number
    number=number-1
print(factorial)
```

```python
#fibonacci series 0 1 1 2 3 5 8 13.....

number=int(input("Enter the number: "))
a,b=0,1
counter = 0
while counter<number:
    print(a)
    c=a+b
    a=b
    b=c
    counter=counter+1
        
```

```python
word=input("Enter the string: ")
counter=len(word)
reverse=""
while counter>=1:
    reverse=reverse+str(word[counter-1])
    
    counter=counter-1

print(reverse)
    
```
