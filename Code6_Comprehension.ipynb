```python
l=[1,2,3,4,45,5]
l1=[]
for i in range(0,len(l)):
    square=l[i] * l[i]
    l1.append(square)

print(l1)
```

```python
l=[1,2,3,4,45,5]
l1=[]
for i in l:
    l1.append(i**2)
print(l1)
```

```python
#similar thing in a shortcut way by using comrehension
[i**2 for i in l]
```

```python
#returning only even number from the list using comprehesnion
[i for i in l if i % 2 ==0]   #syntax [result forLoop Condition]
```
