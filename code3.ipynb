```python
#we have to convert every letter to uppercase
l=["sudh","kumar","pwskills","course"]
l1=[]
for i in l:
    l1.append(i.upper())
print(l1)
```

```python
#we have to separate int or float to a list and string to a different list 
l=[1,2,3,4,4,"sudh","kumar",324,34.456,"abc"]
l1=[]
l2=[]
for i in l:
    if type(i)==int:
        l1.append(i)
    elif type(i)==float:
        l1.append(i)
    else:
        l2.append(i)
print(l1)
print(l2)
```
